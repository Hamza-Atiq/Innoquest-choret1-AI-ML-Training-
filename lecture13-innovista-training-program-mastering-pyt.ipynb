```python
x = torch.tensor([[1.] , [2.] , [3.] , [4.]] , dtype = torch.float32)
y = torch.tensor([[2.] , [4.] , [6.] , [8.]] , dtype = torch.float32)
```

```python
w = torch.tensor(0.0 ,requires_grad = True, dtype = torch.float32)
w
```

```python
import torch.nn as nn
```

```python
def forward(x):
    return w*x
```

```python
loss = nn.MSELoss()
learning_rate = 0.01
optimizer = torch.optim.SGD([w] , learning_rate )
```

```python
num_epochs = 10
for epoch in range(num_epochs):
    y_pred = forward(x)
    loss_true = loss(y_pred , y)

    loss_true.backward()

    optimizer.step()
    optimizer.zero_grad()

    print(f'epoch {epoch} , loss = {loss_true}')
```

# Custom model

```python
class linearregression(nn.Module):

    def __init__(self , input_size , output_size):
        super(linearregression , self).__init__()
        self.linear = nn.Linear(input_size , output_size)

    def forward(self , x):
        output =  self.linear(x)
        return output 


model = linearregression(n_features , output_size  = 1)
```

```python
loss = nn.MSELoss()
learning_rate = 0.01
optimizer = torch.optim.SGD(model.parameters() , learning_rate )
```

```python
num_epochs = 10
for epoch in range(num_epochs):
    y_pred = model(tensor_x)
    loss_true = loss(y_pred , tensor_y)

    loss_true.backward()

    optimizer.step()
    optimizer.zero_grad()

    print(f'epoch {epoch+1} , loss = {loss_true.item()}')
```

# Custom data Loader

```python
import torchvision
import math
from torch.utils.data import DataLoader , Dataset , random_split
```

```python
class Wine(Dataset):

    def __init__(self):

        dataset_wine =  np.loadtxt('/kaggle/input/for-pytorch/wine.csv' , delimiter = ',' , skiprows = 1 , dtype = np.float32)
        self.x = torch.from_numpy(dataset_wine[: , 1:])
        self.y = torch.from_numpy(dataset_wine[: , 0]).long()-1
        self.n_samples = dataset_wine.shape[0] 

        self.x = (self.x - self.x.mean(dim = 0)) / self.x.std(dim = 0)

    def __getitem__(self , index):
        return self.x[index] , self.y[index]

    def __len__(self):
        return self.n_samples

dataset = Wine()

train_size = int(0.8 * len(dataset))
test_size = len(dataset) - train_size

train_dataset , test_dataset = random_split(dataset , [train_size , test_size ])
         
train_loader = DataLoader(dataset = train_dataset , batch_size = 4 , shuffle = True)
test_loader = DataLoader(dataset = test_dataset , batch_size = 4 , shuffle = False)
    
```

```python
dataiter = iter(train_loader)
data = next(dataiter)
features , labels = data
print(features , labels)
```

```python
epochs = 20
total_samples = len(dataset)
n_iteration = math.ceil(total_samples / 4)
print(n_iteration)
```

```python
class logisticregression(nn.Module):

    def __init__(self , input_size , output_size):
        super(logisticregression , self).__init__()
        self.linear = nn.Linear(input_size , output_size)

    def forward(self , x):
        output =  self.linear(x)
        return output 


model = logisticregression(input_size = 13 , output_size  = 3)
```

```python
loss = nn.CrossEntropyLoss()
learning_rate = 0.01
optimizer = torch.optim.SGD(model.parameters() , learning_rate)
```

```python
num_epochs = 20
for epoch in range(num_epochs):
    for i , (inputs , labels) in enumerate(train_loader):
        
        y_pred = model(inputs)
        loss_true = loss(y_pred , labels)
    
        loss_true.backward()
    
        optimizer.step()
        optimizer.zero_grad()

    if epoch % 1 == 0:

        print(f'epoch {epoch+1} , loss = {loss_true.item()}')
```

```python
total_correct_test = 0
total_test_samples = 0
with torch.no_grad():
    for inputs , labels in test_loader:
        y_pred_test = torch.argmax(model(inputs) , dim = 1)
        total_correct_test += (y_pred_test == labels).sum().item()
        total_test_samples += labels.size(0)

    test_accuracy = total_correct_test / total_test_samples

#print(f"Train Accuracy: {train_accuracy:.4f}")
print(f"Test Accuracy: {test_accuracy:.4f}")
```
