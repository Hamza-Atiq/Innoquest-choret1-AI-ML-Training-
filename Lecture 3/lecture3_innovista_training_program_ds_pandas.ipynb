# deep copy

```python
name = 'hamza'
print(name)
name1 = name
print(name1)
name1 = name1.replace('hamza' , 'Atiq')
print(name1)
print(name)
```

# shallow copy

```python
ls1 = list(name)
print(ls1)
ls2 = ls1
print(ls2)
ls2[0] = 'A'
print(ls2)
print(ls1)
ls3 = ls1[:]
print(ls3)
ls3[0] = 'B'
print(ls3)
print(ls1)
```
